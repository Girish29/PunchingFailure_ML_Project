```python
data = data[[' CylindricalConcreteCompressiveStrength','YieldStrength','effective depth','a','a/d', 'c/d','Rho','Cmax_Cmin_Ratio','P_test']]
```

```python
data.columns
```

```python
input = data[[' CylindricalConcreteCompressiveStrength','YieldStrength','effective depth','a','a/d', 'c/d','Rho','Cmax_Cmin_Ratio']]
output = data['P_test']
```

```python
input
```

```python
sns.heatmap(input.corr(), annot=True)
```

```python
input.rename(columns = {' CylindricalConcreteCompressiveStrength':'fc','YieldStrength':'fy','effective depth':'d'},inplace = True)
input
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
# prompt: scale the input and output data

scaler_input = preprocessing.StandardScaler()
scaler_output = preprocessing.StandardScaler()

input_scaled = scaler_input.fit_transform(input)
output_scaled = scaler_output.fit_transform(output.values.reshape(-1,1))
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_scaled,output_scaled,test_size = 0.15, random_state = 1)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    Dense(64, activation = 'relu'),
    Dense(32, activation = 'relu'),
    Dense(1)]
)
```

```python
model.compile(optimizer = 'adam',loss = 'mean_squared_error')
```

```python
model.fit(x = x_train, y = y_train, epochs = 200,batch_size = 64,validation_data = (x_test,y_test))
```

```python
pd.DataFrame(model.history.history)
```

```python
predictions = model.predict(x_test)
predictions
```

```python
predictions_unscaled = scaler_output.inverse_transform(predictions)
```

```python
predictions_unscaled
```

```python
df1 = pd.DataFrame({'P_test': scaler_output.inverse_transform(y_test).flatten(), 'Predictions': predictions_unscaled.flatten()})
df1['error'] = abs(df1['P_test'] - df1['Predictions'])/df1['P_test']*100
df1
```
