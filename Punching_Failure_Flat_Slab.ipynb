```python
from sklearn import tree
regressor = DecisionTreeRegressor(random_state = 42, max_depth = 3)
regressor.fit(x_train,y_train)
fig = plt.figure(figsize=(100,50))
tree.plot_tree(regressor, feature_names = columns,class_names=['P_test'],filled = True)
# plt.savefig('decision_tree.png')
```

```python
def tune_maxDepth(a):
  model = DecisionTreeRegressor(random_state = 42, max_depth = a)
  model.fit(x_train,y_train)
  print(model.predict([[35,450,100,1,0.41,0,5,1.8]])[0])
  print("Score for maxDepth = ",a)
  print("Training Score = ",model.score(x_train,y_train))
  print("Test Score = ",model.score(x_test,y_test))
  return [a,1-model.score(x_train,y_train),1-model.score(x_test,y_test)]
```

```python
data = []
for i in range (1,20):
  data.append(tune_maxDepth(i))
```

```python
df0 = pd.DataFrame(data,columns = ['max_depth','TrainingError', 'TestError'])
df_melt = df0.melt('max_depth', var_name = 'TypeOfError', value_name = 'error')
sns.relplot(data = df_melt, x = 'max_depth', y = 'error', hue = 'TypeOfError', kind = 'line')
```

```python
def tune_min_samples_split(a):
  model = DecisionTreeRegressor(random_state = 42, min_samples_split = a)
  model.fit(x_train,y_train)
  print("Score for min_samples_split = ",a)
  print("Training Score = ",model.score(x_train,y_train))
  print("Test Score = ",model.score(x_test,y_test))
  return [a,1-model.score(x_train,y_train),1-model.score(x_test,y_test)]
```

```python
data = []
for i in range (2,20):
  data.append(tune_min_samples_split(i))
```

```python
df0 = pd.DataFrame(data,columns = ['minSamplessplit','TrainingError', 'TestError'])
df_melt = df0.melt('minSamplessplit', var_name = 'TypeOfError', value_name = 'error')
sns.relplot(data = df_melt, x = 'minSamplessplit', y = 'error', hue = 'TypeOfError', kind = 'line')
```

```python
def tune_min_samples_leaf(a):
  model = DecisionTreeRegressor(random_state = 42, min_samples_leaf = a)
  model.fit(x_train,y_train)
  print("Score for min_samples_leaf = ",a)
  print("Training Score = ",model.score(x_train,y_train))
  print("Test Score = ",model.score(x_test,y_test))
  return [a,1-model.score(x_train,y_train),1-model.score(x_test,y_test)]
```

```python
def tune_nEstimator(a):
  model = XGBRegressor(n_estimators = a)
  model.fit(x_train,y_train.values.ravel())
  print("Score for min_samples_leaf = ",a)
  print("Training Score = ",model.score(x_train,y_train))
  print("Test Score = ",model.score(x_test,y_test))
  return [a,1-model.score(x_train,y_train),1-model.score(x_test,y_test)]

data = []
for i in range(1,20):
  data.append(tune_nEstimator(i))
df = pd.DataFrame(data,columns = ['nEstimator','TrainingError', 'TestError'])
df_melt = df.melt('nEstimator', var_name = 'TypeOfError', value_name = 'error')
sns.relplot(data = df_melt, x = 'nEstimator', y = 'error', hue = 'TypeOfError', kind = 'line')
```

```python
def tune_max_Depth(a):
  model = XGBRegressor(max_depth = a)
  model.fit(x_train,y_train.values.ravel())
  print("Score for maxDepth = ",a)
  print("Training Score = ",model.score(x_train,y_train))
  print("Test Score = ",model.score(x_test,y_test))
  return [a,1-model.score(x_train,y_train),1-model.score(x_test,y_test)]

data = []
for i in range(1,20):
  data.append(tune_max_Depth(i))
df = pd.DataFrame(data,columns = ['max_Depth','TrainingError', 'TestError'])
df_melt = df.melt('max_Depth', var_name = 'TypeOfError', value_name = 'error')
sns.relplot(data = df_melt, x = 'max_Depth', y = 'error', hue = 'TypeOfError', kind = 'line')
```
